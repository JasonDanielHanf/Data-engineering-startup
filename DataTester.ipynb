This jupyter notebook uses the Black Friday Dataset from [Kaggle](https://www.kaggle.com/abhisingh10p14/black-friday?select=train.csv).

Some key concepts utilized within this notebook include

- Importing CSV training dataset
- Converting dataset to a pandas dataframe
- Performing statistical checks:
    - Mean
    - Mode
    - Median
    - Standard Deviation
- Plotting graphs with the relevant data.
- Creating a data quality and data profiling report




```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
```

## Read the CSV and placing it into a Pandas dataframe

```python
data = pd.read_csv("test.csv") 
data.head()

```

## statistical checks

```python
median = data.median()
mean = data.mean()
std_dev = data.std()

median_col = median.values.tolist()
mean_col = mean.values.tolist()
std_dev_col = std_dev.tolist()

headings = ['User_ID','Occupation','Marital Status','Product_Category_1','Product_Category_2','Product_Category_3']

out = pd.DataFrame({'Column Names':headings,'Mean':mean_col,'Median':median_col,'Standard Deviation':std_dev_col})
out.style.set_properties(**{'background-color': 'white','color': 'black'})
```
